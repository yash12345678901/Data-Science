```python
x = data.drop(60, axis = 1)
y = data[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first = True)
y.sample(5)
```

```python
### Splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 1)
```

#### Using Deep Learning Method

### Model Dropout Model

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
        keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
        keras.layers.Dense(30, activation = 'relu'),
        keras.layers.Dense(15, activation = 'relu'),
        keras.layers.Dense(1, activation = 'sigmoid')
])
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 100, batch_size = 8)
```

```python
model.evaluate(x_test, y_test)
```

### Here train accuracy >> test accuracy

```python
y_predict = model.predict(x_test).reshape(-1)
print(y_predict[:10])
```

```python
## round the value
y_pred = np.round(y_predict)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

### Model with dropout layer


```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation = 'sigmoid')
    
])
modeld.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
modeld.fit(x_train, y_train, epochs = 100, batch_size = 8)
```

```python
modeld.evaluate(x_test, y_test)
```

### Now test accuracy is good

```python
y_prediction = modeld.predict(x_test).reshape(-1)
print(y_prediction[:10])
```

### round the value to nearest integer ie 0 or 1

```python
y_prediction = np.round(y_prediction)
print(y_prediction[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_prediction))
```
